```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

veri = {
    'Sistem_Tipi_IMA_CN':   [1, 0, 0, 1, 0, 1, 0, 0, 1, 0],
    'Sistem_Tipi_IMA_LCDS': [0, 1, 0, 0, 1, 0, 1, 1, 0, 1],
    'Agirlik_Gram': [500, 1200, 300, 800, 150, 2000, 100, 1500, 450, 3000],
    'Uretim_Suresi_Dk': [45, 90, 30, 60, 15, 180, 10, 120, 40, 240],
    'Satis_Fiyati_TL': [4500, 12000, 3100, 5400, 1800, 15000, 1500, 9500, 4100, 22000]
}

df = pd.DataFrame(veri)

X = df[['Sistem_Tipi_IMA_CN', 'Sistem_Tipi_IMA_LCDS', 'Agirlik_Gram', 'Uretim_Suresi_Dk']]
y = df['Satis_Fiyati_TL']

model = LinearRegression()
model.fit(X, y)

def predict_sales_price(system_ima_cn, system_ima_lcds, weight_gram, production_minutes):
    X_new = np.array([[system_ima_cn, system_ima_lcds, weight_gram, production_minutes]])
    return float(model.predict(X_new)[0])

print("✅ Yapay zeka modeli başarıyla eğitildi!")
print("R² skor:", r2_score(y, model.predict(X)))

tahmin_edilen_fiyat = predict_sales_price(0, 1, 1000, 75)
print(f"🔮 Bu sistem için önerilen satış fiyatı: {tahmin_edilen_fiyat:.2f} TL")
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

veri = {
    'Sistem_Tipi_IMA_CN':   [1, 0, 0, 1, 0, 1, 0, 0, 1, 0],
    'Sistem_Tipi_IMA_LCDS': [0, 1, 0, 0, 1, 0, 1, 1, 0, 1],
    'Agirlik_Gram': [500, 1200, 300, 800, 150, 2000, 100, 1500, 450, 3000],
    'Uretim_Suresi_Dk': [45, 90, 30, 60, 15, 180, 10, 120, 40, 240],
    'Satis_Fiyati_TL': [4500, 12000, 3100, 5400, 1800, 15000, 1500, 9500, 4100, 22000]
}

df = pd.DataFrame(veri)

X = df[['Sistem_Tipi_IMA_CN', 'Sistem_Tipi_IMA_LCDS', 'Agirlik_Gram', 'Uretim_Suresi_Dk']]
y = df['Satis_Fiyati_TL']

model = LinearRegression()
model.fit(X, y)

def predict_sales_price(system_ima_cn, system_ima_lcds, weight_gram, production_minutes):
    X_new = np.array([[system_ima_cn, system_ima_lcds, weight_gram, production_minutes]])
    return float(model.predict(X_new)[0])

print("✅ Yapay zeka modeli başarıyla eğitildi!")
print("R² skor:", r2_score(y, model.predict(X)))

tahmin_edilen_fiyat = predict_sales_price(0, 1, 1000, 75)
print(f"🔮 Bu sistem için önerilen satış fiyatı: {tahmin_edilen_fiyat:.2f} TL")
```
